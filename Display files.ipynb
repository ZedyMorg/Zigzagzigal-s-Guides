```python
import pandas as pd
import seaborn as sns
```

```python
ach_guide = pd.read_csv('guide_with_achievements.csv')
ach_guide.drop(ach_guide.columns[0], axis=1, inplace=True)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
def make_clickable(val):
    return f'<a target="_blank" href="{val}">{val}</a>'

def remove_trailing_zeros(val):
    if pd.notna(val):
        return round(val,2)
    
def color_achieved(val):
    green = 'background-color: green'
    red = 'background-color: red'
    
    if (val == True) & (type(val) == bool):
        color = green
        return color
    
    if (val == False) & (type(val) == bool):
        color = red
        return color

    
cm = sns.color_palette("crest", as_cmap=True)
```

```python
not_ach_GS = ach_guide[(ach_guide['achieved'] == False)]
not_ach_GS = not_ach_GS[pd.isna(not_ach_GS['Link']) | not_ach_GS['Country'].str.contains('GS')]
```

```python
#Preparing dictonnary of functions to apply to display our dataframe

format_dict = {'Link': make_clickable, 'precision' : 3}

trailing_zero_cols = ['Culture', 'Diplomacy', 'Domination', 'Religion', 'Science', 'Flexibility']

for i in trailing_zero_cols:
    format_dict[i] = remove_trailing_zeros
```

```python
show_df = not_ach_GS.copy()

user_choice = '-1'
filters = []

print('What kind of civilization you want to play? \nEnter 0 to confirm selection. \n\nEnter 1 for Flexible. \nEnter 2 for Culture. \nEnter 3 for Diplomacy \nEnter 4 for Domination \nEnter 5 for Religion \nEnter 6 for Science\n')
print('The order of your choices will be the order of importance in the filter in case of egality')
while user_choice != '0':
    user_choice = input()
    
    if user_choice == '1':
        filters.append('Flexibility')
        
    if user_choice == '2':
        filters.append('Culture')
        
    if user_choice == '3':
        filters.append('Diplomacy')
        
    if user_choice == '4':
        filters.append('Domination')
        
    if user_choice == '5':
        filters.append('Religion')
        
    if user_choice == '6':
        filters.append('Science')
        
display(show_df.sort_values(filters, ascending = False).reset_index(drop = True).style.background_gradient(cmap=cm).applymap(color_achieved).format(format_dict)) 
```
